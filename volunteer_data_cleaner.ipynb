```python
# Make a deliberately messy volunteer sign-up list to practice on
import pandas as pd

data = {
    "name":  ["Maria Lopez","maria lopez","James Chen","J. Chen ","Aisha Khan","Tom Reed","tom reed","", "Priya Patel","David Kim"],
    "email": ["maria@x.com","maria@x.com","james@x.com","james@x.com","aisha@x.com","tom@x.com","tom@x.com","ghost@x.com","priya@x.com",""],
    "signup_date": ["2026-06-01","2026-06-01","06/02/2026","06/02/2026","2026-06-03","June 4 2026","June 4 2026","2026-06-05","2026-06-06","2026-06-07"],
    "hours": [4, 4, 6, 6, "", 3, 3, 2, 8, 5],
}
df = pd.DataFrame(data)
df.to_csv("messy_volunteers.csv", index=False)
print("Created messy_volunteers.csv")
```

```python
df = pd.read_csv("messy_volunteers.csv")
print(df)        # see the whole thing
print()
df.info()        # see column types and where values are missing
```

```python
# --- CLEAN THE DATA ---

# 1. Tidy the names: remove extra spaces, fix the capitalization
df["name"] = df["name"].str.strip().str.title()

# 2. Tidy the emails: remove spaces, make them all lowercase
df["email"] = df["email"].str.strip().str.lower()

# 3. Fix the dates: turn all three formats into one clean date
df["signup_date"] = pd.to_datetime(df["signup_date"], format="mixed")

# 4. Remove duplicate people. Same EMAIL = same person, even if the
#    name is typed differently. Keep the first time they appear.
df = df.drop_duplicates(subset="email", keep="first")

# 5. Drop rows that have no name at all (the junk "ghost" row)
df = df.dropna(subset=["name"])

# 6. Fill in missing hours with 0 so we can add them up later
df["hours"] = df["hours"].fillna(0)

# Reset the row numbers and show the result
df = df.reset_index(drop=True)
print(df)
```

```python
# --- SUMMARIZE: answer a real question from the data ---

# Q1: How many total volunteer hours did we get, and from how many people?
total_hours = df["hours"].sum()
total_people = len(df)
print(f"{total_people} volunteers contributed {total_hours:.0f} total hours.")
print(f"Average: {total_hours / total_people:.1f} hours per volunteer.")
print()

# Q2: How many sign-ups per day? (groupby = bucket rows, then count)
per_day = df.groupby("signup_date").size()
print("Sign-ups per day:")
print(per_day)
```

```python
# --- SAVE & DOWNLOAD the cleaned file ---
df.to_csv("clean_volunteers.csv", index=False)
print("Saved clean_volunteers.csv")

from google.colab import files
files.download("clean_volunteers.csv")   # downloads to your computer
```
